```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np


df = pd.read_csv('pinguinos.csv')
df = df.dropna(subset=['sex'], axis= 'index')
sns.boxplot(df, x='sex', y= "body_mass_g"). set(title = "body_mass_g vs sexo")
plt.show()
```

```python
# Renombrar columnas a español
df_espanol = df.rename(columns={'culmen_length_mm':'largo_culmen_mm',
                               'culmen_depth_mm':'profundidad_culmen_mm',
                               'flipper_length_mm':'largo_aleta_mm',
                               'body_mass_g':'masa_corporal_g'})

# Calcular la matriz de correlación
correlation_matrix = df_espanol.corr(numeric_only=True)

# Crear una máscara para la parte triangular superior (incluyendo la diagonal)
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))

# Añadir asteriscos a las correlaciones fuertes
annot_matrix = correlation_matrix.copy()
for i in range(len(annot_matrix.columns)):
    for j in range(len(annot_matrix.columns)):
        if abs(annot_matrix.iloc[i, j]) > 0.7 and i <= j: # Check only upper triangle including diagonal
             annot_matrix.iloc[i, j] = f"{annot_matrix.iloc[i, j]:.2f}*"
        else:
            annot_matrix.iloc[i, j] = f"{annot_matrix.iloc[i, j]:.2f}"


# Crear un mapa de calor (heatmap) con valores y colores
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=annot_matrix, cmap='coolwarm', fmt="", mask=mask)
plt.title('Matriz de Correlación (Triangular Inferior con Correlaciones Fuertes)')
plt.show()
```

```python
# Mostrar las correlaciones con masa corporal
display(correlation_matrix[['masa_corporal_g']])
```

## Histogramas con campana de gauss

### Subtask:
Analizar la distribución de variables numéricas y superponer una curva de distribución normal (campana de Gauss) para ver qué tan cerca se ajustan los datos a una distribución normal.


**Reasoning**:
Create histograms with superimposed normal distribution curves for the numerical columns to analyze their distributions.



```python
numerical_cols = ['culmen_length_mm', 'culmen_depth_mm', 'flipper_length_mm', 'body_mass_g']

for col in numerical_cols:
    plt.figure(figsize=(8, 6))
    sns.histplot(data=df, x=col, kde=False, bins=20, stat="density")

    # Calculate mean and standard deviation
    mean = df[col].mean()
    std = df[col].std()

    # Generate values for the normal distribution curve
    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 100)
    p = 1/(np.sqrt(2*np.pi)*std) * np.exp(-(x-mean)**2/(2*std**2))

    # Plot the normal distribution curve
    plt.plot(x, p, 'k', linewidth=2)

    plt.title(f'Distribución de {col} con curva normal')
    plt.xlabel(col)
    plt.ylabel('Densidad')
    plt.show()
```
